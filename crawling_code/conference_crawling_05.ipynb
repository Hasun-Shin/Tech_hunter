```python
# 함수로 만들기
def get_articles():
    
    import requests
    from bs4 import BeautifulSoup
    import re
  
    url = "http://conference.etnews.com/index.html?page=1"
    
    exchange = soup.select('body > div > main > div > ul ')
    
    
    
    #제목
    import pandas as pd

    title_list = []
    exchange_title = soup.select('li > a > dl')

    for titles in exchange_title:
        title = titles.text.strip()
        title_list.append(title)
    title =pd.DataFrame(title_list, columns=['title'])
    
    
    
    #링크만 뽑아내기
   
    link_list = []
    exchange_link = soup.select('li > a ')

    for links in exchange_link:
        a = "http://conference.etnews.com/"
        link = (a+links.attrs['href'])
        link_list.append(link)
    link =pd.DataFrame(link_list, columns=['link'])
    
    
    
    
    
    #이미지 뽑아내기
    img_list = []
    exchange_img = soup.select('body > div > main > div > ul > li > a > dl > dd.thumb > img')

    for imgs in exchange_img:
    
        imgs = imgs.attrs['src']
        img_list.append(imgs)


    img =pd.DataFrame(link_list, columns=['img'])
    
    result=pd.concat([title, link], axis=1)
     
    return result

 
```
