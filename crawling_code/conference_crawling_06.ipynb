```python
# -*- coding: utf-8 -*-
import requests
from bs4 import BeautifulSoup
import re
```

```python


url = "http://www.bloter.net/archives/category/bloterconference"
req = requests.get(url).text
soup = BeautifulSoup(req,'html.parser') 

exchange = soup.select('#category--body__container > div ')


```

```python
#제목,일시,장소



for titles in exchange :
    #print(titles.attrs['title'])
    #title = titles.text
    

    print(titles )
```

```python
import pandas as pd

title_list = []
exchange_tilte = soup.select('body > div > main > div > ul > li > a > dl  ')

for titles in exchange:
    title = titles.text.strip()
    title_list.append(title)
title =pd.DataFrame(title_list, columns=['title'])
title
```

```python
# 하이퍼링크

exchange = soup.select('body > div > main > div > ul > li > a')
for link in exchange :

    a = "http://conference.etnews.com/"

    print(a+link.attrs['href'])
    #print(link)
```

```python
#링크 데이터프레임

import pandas as pd

link_list = []
#exchange_link = soup.select('body > div > main > div > ul > li > a  ')

for links in exchange:
    a = "http://conference.etnews.com/"
    link = (a+links.attrs['href'])
    link_list.append(link)
link =pd.DataFrame(link_list, columns=['link'])
link
```

```python
# 이미지

exchange = soup.select('body > div > main > div > ul > li > a > dl > dd.thumb > img')
for img in exchange :

    

    print(img.attrs['src'])
    #print(link)
```

```python
#이미지 데이터프레임

import pandas as pd

img_list = []
exchange_img = soup.select('body > div > main > div > ul > li > a > dl > dd.thumb > img')

for imgs in exchange_img:
    
    imgs = imgs.attrs['src']
    img_list.append(imgs)


img =pd.DataFrame(link_list, columns=['img'])
img
```

```python
result = pd.concat([title, link, img], axis=1) #axis=1 은 열 !!! 
result
```
